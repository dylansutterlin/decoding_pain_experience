```python
import scipy

root_dir = r'C:\Users\Dylan\Desktop\UM_Bsc_neurocog\E22\Projet_Ivado_rainvillelab\pipeline_MVPA\dot_product'
mat_SPM = scipy.io.loadmat(os.path.join(root_dir,'dot_SPM_Nshocks_NPS.mat'),simplify_cells=True)
# = pd.DataFrame(mat_result_SPM_NPS)
result_SPM_NPS = pd.concat([pd.DataFrame(mat_SPM['nps_values']),pd.DataFrame(mat_SPM['image_names']),pd.DataFrame(mat_SPM['data_objects'])], axis = 1)
result_SPM_NPS.columns = ['nps_values', 'image_names','data_objects']
```

#### Descriptive statistics and plots

```python
pd.to_numeric(result_SPM_NPS['nps_values']).describe()
```

```python
import plotly.express as px

fig = px.violin(result_SPM_NPS, y='nps_values', box=True, # draw box plot inside the violin
                points='all', # can be 'outliers', or False
                title='Dot product results for Nshocks SPM-NPS'
               )
fig.show()
```

#### T-test to zero


```python
#T-test
from scipy import stats

mu=0
t_value,p_value=stats.ttest_1samp(pd.to_numeric(result_SPM_NPS['nps_values']),mu)

one_tailed_p_value=float("{:.6f}".format(p_value/2)) # Since alternative hypothesis is one tailed, We need to divide the p value by 2.

print('Test statistic is %f'%float("{:.6f}".format(t_value)))
print('P value is %f'%float("{:.6f}".format(p_value)))
```
